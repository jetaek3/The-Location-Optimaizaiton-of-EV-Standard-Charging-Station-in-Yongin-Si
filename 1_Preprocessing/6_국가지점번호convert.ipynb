# 위도경도 좌표값 WSG-84 -> UTM-K

```python
import pandas as pd
import geopandas as gpd
import pickle
from pyproj import Proj, transform
```

```python
df = pd.read_csv('/Users/jacob/Desktop/VS_Code_Repository/DeepLearning_Alpaco/빅콘테스트2022/Raw data/ev_app_resident.csv')
```

```python
proj_UTMK = Proj(init='epsg:5178')
proj_WSG84 = Proj(init='epsg:4633')
```

```python
def national_geocode(df):
    global proj_UTMK
    global proj_WSG84

    dic_x={'7':'가','8':'나','9':'다','10':'라','11':'마','12':'바','13':'사'}
    dic_y={'13': '가','14': '나','15': '다','16': '라','17': '마','18': '바','19': '사','20': '아'}

    df['UTM-K_longitude'],df['UTM-K_latitude']= transform(proj_WSG84,proj_UTMK,df['cell_xcrd'],df['cell_ycrd'])
    df['UTM-K_longitude']=df['UTM-K_longitude'].apply(int).apply(str)
    df['UTM-K_latitude']=df['UTM-K_latitude'].apply(int).apply(str)
    
    print(df.info)
    print(df.describe)
    df['국가지점번호']=df['UTM-K_longitude'].str[:-5]+df['UTM-K_latitude'].str[:-5]+df['UTM-K_longitude'].str[-5:-2]+df['UTM-K_latitude'].str[-5:-2]
    df['국가지점번호']=pd.Series([dic_x[i[:-8]]+dic_y[i[-8:-6]]+i[-6:] for i in df['국가지점번호']])

    return df
```

```python
national_geocode(df)
```

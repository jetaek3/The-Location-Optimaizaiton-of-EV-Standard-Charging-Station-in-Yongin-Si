# Geopandas 행정동 추출

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, Polygon, LineString
from tqdm import tqdm
```

```python
sdf = gpd.read_file('/Users/jacob/Desktop/VS_Code_Repository/DeepLearning_Alpaco/빅콘테스트2022/Raw data/용인시_읍면동_EPSG4326.geojson')
df = gpd.read_file('/Users/jacob/Desktop/VS_Code_Repository/DeepLearning_Alpaco/빅콘테스트2022/Raw data/용인시_Grid_filtered_v2(dropped_covered).geojson')
sdf.to_crs('epsg:5178',inplace=True)
```

## 행정동 geometry 안에 격자 centroid point 들어가면 읍면동 column 추가

```python
for i,irow in tqdm(df.iterrows(),total=df.shape[0]):
    for j,jrow in sdf.iterrows():
        if df.geometry[i].within(sdf.geometry[j]):
            df.loc[i,'읍면동']=sdf.loc[j,'adm_nm'].split()[2]
```

## 읍면동 -> 전기차 등록 대수 변환

```python
df['ev_cnt']=df['읍면동']
df.replace({"ev_cnt":dic},inplace=True)
```
